```python
import pandas as pd
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
```

```python
X = data.iloc[:, 0:11]
y = data.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
X.isnull().sum()
```

```python
X.dtypes
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 32, stratify= y)    #startify means test size follows the distribution of given parameter. 
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X_train = ss.fit_transform(X_train)
```

```python
X_train = pd.DataFrame(X_train)
```

```python
X_train.describe()
```

```python
X_train.columns = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
```

```python
X_train.head(20)
```

```python
X_test = ss.transform(X_test)
```

```python
X_test = pd.DataFrame(X_test)
```

```python
X_test.columns = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
```

```python
X_test.head(20)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
dtc = DecisionTreeClassifier(random_state= 1)
rfc = RandomForestClassifier(random_state = 1)
gbc = GradientBoostingClassifier(random_state= 1)
lr = LogisticRegression(random_state= 1)
svc = SVC()
knn = KNeighborsClassifier()
sgd = SGDClassifier()
gnb = GaussianNB()
```

```python
dtc.fit(X_train, y_train)
rfc.fit(X_train, y_train)
gbc.fit(X_train, y_train)
lr.fit(X_train, y_train)
svc.fit(X_train, y_train)
knn.fit(X_train, y_train)
sgd.fit(X_train, y_train)
gnb.fit(X_train, y_train)
```

```python
print(dtc.score(X_test, y_test))
print(rfc.score(X_test, y_test))
print(gbc.score(X_test, y_test))
print(lr.score(X_test, y_test))
print(svc.score(X_test, y_test))
print(knn.score(X_test, y_test))
print(sgd.score(X_test, y_test))
print(gnb.score(X_test, y_test))
```

```python
from sklearn.metrics import plot_confusion_matrix , f1_score
```

```python
plot_confusion_matrix(dtc, X_test, y_test)
plot_confusion_matrix(rfc, X_test, y_test)
plot_confusion_matrix(gbc, X_test, y_test)
plot_confusion_matrix(lr, X_test, y_test)
plot_confusion_matrix(svc, X_test, y_test)
plot_confusion_matrix(knn, X_test, y_test)
plot_confusion_matrix(sgd, X_test, y_test)
plot_confusion_matrix(gnb, X_test, y_test)
```

```python
sns.scatterplot(data = data, x= X_train['alcohol'], y= X_train['chlorides'])
```

```python
c = sns.color_palette('Set2', 6)
sns.scatterplot(data = data, x= X_train['alcohol'], y= X_train['chlorides'], hue = y, palette= c)
```

```python
y_pred_dtc = dtc.predict(X_test)
y_pred_rfc = rfc.predict(X_test)
y_pred_gbc = gbc.predict(X_test)
y_pred_lr = lr.predict(X_test)
y_pred_svc = svc.predict(X_test)
y_pred_knn = knn.predict(X_test)
y_pred_sgd = sgd.predict(X_test)
y_pred_gnb = gnb.predict(X_test)
```

```python
print(f1_score(y_test, y_pred_dtc, average= 'micro'))
print(f1_score(y_test, y_pred_rfc, average= 'micro'))
print(f1_score(y_test, y_pred_gbc, average= 'micro'))
print(f1_score(y_test, y_pred_lr, average= 'micro'))
print(f1_score(y_test, y_pred_svc, average= 'micro'))
print(f1_score(y_test, y_pred_knn, average= 'micro'))
print(f1_score(y_test, y_pred_sgd, average= 'micro'))
print(f1_score(y_test, y_pred_gnb, average= 'micro'))
```
