```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()
```

```python
y_km = km.fit_predict(data)
```

```python
y_km[0:100]
```

```python
obj = []

for k in range(1,10):
  km = KMeans(n_clusters= k)
  km.fit(data)
  obj.append(km.inertia_)
```

```python
obj
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,10), obj, marker= 'o')
plt.xlabel(k)
plt.ylabel('obj')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(data, y_km)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'n_clusters': [3,4,5,6,7,8], 'init': ['k-means++', 'random'], 'n_init': [8,9,11,12,13], 'tol': [0.00001,0.0001,0.001,0.01,0.1]}
```

```python
gscv = GridSearchCV(KMeans(), param_grid= params)
```
